```python
import re
import pysearch
from pysearch import methods_path, applications_path, problems_path

class Parser:
    def __init__(self):
        self.problem_regex = re.compile(r'Problem: (.+)')
        self.method_regex = re.compile(r'Method: (.+)')
        self.application_regex = re.compile(r'Application: (.+)')
    
    def parse(self, text):
        tags = []
        for line in text.split('\n'):
            # check if the line matches any of the regular expressions
            problem_match = self.problem_regex.search(line)
            if problem_match:
                tags.append(('Problem', problem_match.group(1)))
                continue
                
            method_match = self.method_regex.search(line)
            if method_match:
                tags.append(('Method', method_match.group(1)))
                continue
                
            application_match = self.application_regex.search(line)
            if application_match:
                tags.append(('Application', application_match.group(1)))
                continue
                
        return tags
```

```python
p = pysearch.pysearch(methods_path, applications_path, problems_path)
text = p.search()

parser = Parser()
tags = parser.parse(text)

for tag in tags:
    print(tag[0], ':', tag[1])
```
